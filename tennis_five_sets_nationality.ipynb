```python

import datetime
import glob
import pandas as pd
import os
import os.path as osp
import sys
```

```python
def readATPMatchesParseTime(dirname):
    """Reads ATP matches and parses time into datetime object"""
    allFiles = glob.glob(dirname + "/atp_matches_" + "????.csv")
    matches = pd.DataFrame()
    container = list()
    for filen in allFiles:
        df = pd.read_csv(filen,
                         index_col=None,
                         header=0,
                         encoding = "ISO-8859-1")
        container.append(df)
    matches = pd.concat(container)
    return matches
    
```

```python
df = readATPMatchesParseTime('tennis_atp')
df.head()
```

```python
# Keep only Grand slam matches
df = df[df['tourney_level'] == 'G']
df.head()
```

```python
# Remove useless columns 
keep = ['winner_name', 'loser_name', 'winner_ioc', 'loser_ioc', 'score']
filtered = df[keep]
filtered.head()
```

```python
# Parse score
def is_five_set(score_string):
    if score_string != score_string:
        print 'Error on score_string: ' + str(score_string)
    sets = score_string.split(' ')
    return len(sets) == 5
```

```python
defects = filtered[filtered['score'] != filtered['score']]
defects
```

```python
filtered = filtered[filtered.score == filtered.score]
```

```python
filtered['is_five_sets'] = filtered['score'].apply(is_five_set)
filtered.head()
```
